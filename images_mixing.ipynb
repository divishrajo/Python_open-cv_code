## Combining images to form a single image

```python
import cv2
import matplotlib.pyplot as plt
```

### Reading the two images "frame.jpg" and "sea.jpg"

```python
frame = cv2.imread("frame.jpg")
sea = cv2.imread("sea.jpg")
```

### Plotting the images 

```python
plt.imshow(frame) # It will plot our frame.png image
```

```python
plt.imshow(sea)   # It will plot our sea.png image
```

### Cropping the images frame and sea to combine

```python
frame_cropped = frame[157:520,419:930]    # Cropping the image frame 
cv2.imshow("cropped_frame_image",frame_cropped)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
sea_cropped = sea[157:520,419:930]   # Cropping the image sea
cv2.imshow("cropped_sea _image",sea_cropped)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Combining the images frame and sea_cropped

```python
frame[157:520,419:930] = sea_cropped   # Combining the cropped images
cv2.imshow("Result",frame)
cv2.waitKey()
cv2.destroyAllWindows()
```
